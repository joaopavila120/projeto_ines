```python
# ============================================================
# ANÁLISE ESTATÍSTICA - DATASET RECOLHA_DE_DADOS
# ============================================================

# -------------------------
# 0) IMPORTS
# -------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import shapiro, ttest_ind, mannwhitneyu, chi2_contingency, fisher_exact

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# -------------------------
# 1) CARREGAR DADOS
# -------------------------
file_path = "data.xlsx"
df = pd.read_excel(file_path, sheet_name="Recolha_de_Dados")

print("Dimensão original:", df.shape)
print("\nColunas originais:")
print(df.columns.tolist())

# -------------------------
# 2) LIMPEZA DAS COLUNAS
# -------------------------
# remover espaços extras dos nomes
df.columns = [str(c).strip() for c in df.columns]

# renomear colunas problemáticas, se existirem
rename_dict = {}

if "" in df.columns:
    rename_dict[""] = "Nome"

if "Unnamed: 18" in df.columns:
    rename_dict["Unnamed: 18"] = "Tipo_intervencao"

if "nº processo" in df.columns:
    rename_dict["nº processo"] = "n_processo"

if "Ventilação" in df.columns:
    rename_dict["Ventilação"] = "Ventilacao"

if rename_dict:
    df = df.rename(columns=rename_dict)

print("\nColunas após limpeza:")
print(df.columns.tolist())

# -------------------------
# 3) CONVERSÕES E PADRONIZAÇÃO
# -------------------------
# datas
date_cols = ["Data de nascimento", "Data de admissao na UCIPed", "Data_alta_UCIPed"]
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors="coerce")

# numéricas
numeric_cols = ["Idade", "Tempo_total_UCIP_dias", "Dias de ventilação", "Dias de sedoanalgesia"]
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# função padronizar sim/não
def normalizar_sim_nao(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip().lower()
    if x in ["sim", "s", "yes", "y"]:
        return "Sim"
    if x in ["não", "nao", "n", "no"]:
        return "Não"
    return str(x).strip()

# padronizar binárias existentes
binary_cols = [
    "Doença psiquiátrica prévia",
    "Ventilacao",
    "Sedo-analgia",
    "Medicação psiquiátrica iniciada",
    "Seguimento psic após alta"
]

for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].apply(normalizar_sim_nao)

# sexo
if "Sexo" in df.columns:
    df["Sexo"] = df["Sexo"].astype(str).str.strip().str.capitalize()

# doença crónica prévia -> Sim/Não sem inferir detalhe clínico
if "Doença crónica PRÉVIA" in df.columns:
    def doenca_cronica_bin(x):
        if pd.isna(x):
            return np.nan
        x = str(x).strip().lower()
        if x in ["não", "nao"]:
            return "Não"
        return "Sim"
    df["Doença_cronica_previa_bin"] = df["Doença crónica PRÉVIA"].apply(doenca_cronica_bin)

print("\nDimensão final:", df.shape)
print("Número total de registos analisados:", len(df))

# -------------------------
# 4) FUNÇÕES AUXILIARES
# -------------------------
def tabela_frequencias(series, dropna=False):
    s = series.copy()
    freq = s.value_counts(dropna=dropna)
    perc = s.value_counts(dropna=dropna, normalize=True) * 100
    return pd.DataFrame({
        "n": freq,
        "%": perc.round(1)
    })

def resumo_continuo(series, nome="Variável"):
    s = pd.to_numeric(series, errors="coerce").dropna()

    if len(s) == 0:
        return pd.DataFrame([{
            "Variável": nome,
            "n": 0,
            "Distribuição": "Sem dados",
            "Resumo": np.nan,
            "Shapiro_p": np.nan,
            "Min": np.nan,
            "Max": np.nan
        }])

    if len(s) < 3:
        return pd.DataFrame([{
            "Variável": nome,
            "n": len(s),
            "Distribuição": "Amostra insuficiente",
            "Resumo": f"Mediana={s.median():.2f} | IIQ=({s.quantile(0.25):.2f}-{s.quantile(0.75):.2f})",
            "Shapiro_p": np.nan,
            "Min": round(s.min(), 2),
            "Max": round(s.max(), 2)
        }])

    p_shapiro = shapiro(s).pvalue
    if p_shapiro > 0.05:
        resumo = f"{s.mean():.2f} ± {s.std(ddof=1):.2f}"
        dist = "Normal"
    else:
        resumo = f"{s.median():.2f} ({s.quantile(0.25):.2f}-{s.quantile(0.75):.2f})"
        dist = "Não normal"

    return pd.DataFrame([{
        "Variável": nome,
        "n": len(s),
        "Distribuição": dist,
        "Resumo": resumo,
        "Shapiro_p": round(p_shapiro, 4),
        "Min": round(s.min(), 2),
        "Max": round(s.max(), 2)
    }])

def barplot_simples(series, titulo, xlabel="", ylabel="Frequência", rotate=0):
    s = series.dropna()
    if len(s) == 0:
        print(f"Sem dados para gráfico: {titulo}")
        return

    order = s.value_counts().index
    plt.figure(figsize=(10, 6))
    sns.countplot(x=s, order=order)
    plt.title(titulo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=rotate, ha="right")
    plt.tight_layout()
    plt.show()

def boxplot_simples(series, titulo, ylabel=""):
    s = pd.to_numeric(series, errors="coerce").dropna()
    if len(s) == 0:
        print(f"Sem dados para gráfico: {titulo}")
        return

    plt.figure(figsize=(7, 5))
    sns.boxplot(y=s)
    plt.title(titulo)
    plt.ylabel(ylabel)
    plt.tight_layout()
    plt.show()

def comparar_continua_2_grupos(df_in, var_continua, var_grupo):
    tmp = df_in[[var_continua, var_grupo]].dropna()
    grupos = list(tmp[var_grupo].unique())

    if len(grupos) != 2:
        return {"teste": "N/A", "p": np.nan}

    g1 = pd.to_numeric(tmp[tmp[var_grupo] == grupos[0]][var_continua], errors="coerce").dropna()
    g2 = pd.to_numeric(tmp[tmp[var_grupo] == grupos[1]][var_continua], errors="coerce").dropna()

    if len(g1) < 2 or len(g2) < 2:
        return {"teste": "N/A", "p": np.nan}

    p1 = shapiro(g1).pvalue if len(g1) >= 3 else 0
    p2 = shapiro(g2).pvalue if len(g2) >= 3 else 0

    if p1 > 0.05 and p2 > 0.05:
        _, p = ttest_ind(g1, g2, equal_var=False, nan_policy="omit")
        return {"teste": "t de Student (Welch)", "p": round(p, 4)}
    else:
        _, p = mannwhitneyu(g1, g2, alternative="two-sided")
        return {"teste": "Mann-Whitney", "p": round(p, 4)}

def comparar_categorica_2_grupos(df_in, var_cat, var_grupo):
    tmp = df_in[[var_cat, var_grupo]].dropna()
    tabela = pd.crosstab(tmp[var_cat], tmp[var_grupo])

    if tabela.shape[1] != 2:
        return {"teste": "N/A", "p": np.nan}

    if tabela.shape == (2, 2):
        _, p = fisher_exact(tabela)
        return {"teste": "Exato de Fisher", "p": round(p, 4)}
    else:
        _, p, _, _ = chi2_contingency(tabela)
        return {"teste": "Qui-quadrado", "p": round(p, 4)}

def interpretar_p(p):
    if pd.isna(p):
        return ""
    return "Significativo" if p < 0.05 else "Não significativo"

# -------------------------
# 5) INÍCIO DA ANÁLISE
# -------------------------
print("\n" + "="*70)
print("1. CARACTERIZAÇÃO GLOBAL DA POPULAÇÃO CONSIDERADA NO FICHEIRO")
print("="*70)

# Aqui, sem inventar total externo: usa apenas n = 38
tabela_global = pd.DataFrame({
    "Descrição": ["Registos analisados na folha Recolha_de_Dados"],
    "n": [len(df)],
    "%": [100.0]
})
print(tabela_global)

plt.figure(figsize=(6, 5))
sns.barplot(data=tabela_global, x="Descrição", y="n")
plt.title("Número total de registos analisados")
plt.ylabel("n")
plt.xlabel("")
plt.tight_layout()
plt.show()

# -------------------------
# 6) 2. CARACTERIZAÇÃO DA AMOSTRA
# -------------------------
print("\n" + "="*70)
print("2. CARACTERIZAÇÃO DA AMOSTRA AVALIADA")
print("="*70)
print(f"Tamanho da amostra: n = {len(df)}")

# 2.1 Sociodemográfica
print("\n--- 2.1 Caracterização Sociodemográfica ---")

if "Idade" in df.columns:
    print("\nResumo da Idade:")
    print(resumo_continuo(df["Idade"], "Idade"))

if "Sexo" in df.columns:
    print("\nDistribuição por Sexo:")
    print(tabela_frequencias(df["Sexo"]))

if "Idade" in df.columns:
    boxplot_simples(df["Idade"], "Boxplot da Idade", ylabel="Idade (anos)")

if "Sexo" in df.columns:
    barplot_simples(df["Sexo"], "Distribuição por Sexo", xlabel="Sexo")

# 2.2 Clínica prévia
print("\n--- 2.2 Caracterização Clínica Prévia ---")

if "Doença psiquiátrica prévia" in df.columns:
    print("\nAntecedentes psiquiátricos:")
    print(tabela_frequencias(df["Doença psiquiátrica prévia"]))

if "Doença_cronica_previa_bin" in df.columns:
    print("\nDoença médica crónica prévia:")
    print(tabela_frequencias(df["Doença_cronica_previa_bin"]))

if "Doença psiquiátrica prévia" in df.columns:
    barplot_simples(df["Doença psiquiátrica prévia"], "Antecedentes psiquiátricos", xlabel="Antecedente psiquiátrico")

if "Doença_cronica_previa_bin" in df.columns:
    barplot_simples(df["Doença_cronica_previa_bin"], "Doença médica crónica prévia", xlabel="Doença crónica")

# 2.3 Internamento na UCIP
print("\n--- 2.3 Caracterização do Internamento na UCIP ---")

if "Categoria_diagnóstico" in df.columns:
    print("\nMotivo de internamento (Categoria_diagnóstico):")
    print(tabela_frequencias(df["Categoria_diagnóstico"]))

if "Tempo_total_UCIP_dias" in df.columns:
    print("\nDuração do internamento:")
    print(resumo_continuo(df["Tempo_total_UCIP_dias"], "Duração do internamento"))

if "Ventilacao" in df.columns:
    print("\nNecessidade de ventilação mecânica:")
    print(tabela_frequencias(df["Ventilacao"]))

if "Dias de ventilação" in df.columns:
    print("\nDuração da ventilação:")
    print(resumo_continuo(df["Dias de ventilação"], "Duração da ventilação"))

if "Dias de sedoanalgesia" in df.columns:
    print("\nDuração da sedoanalgesia:")
    print(resumo_continuo(df["Dias de sedoanalgesia"], "Duração da sedoanalgesia"))

if "Categoria_diagnóstico" in df.columns:
    barplot_simples(df["Categoria_diagnóstico"], "Motivo de internamento", xlabel="Categoria diagnóstica", rotate=30)

if "Tempo_total_UCIP_dias" in df.columns:
    boxplot_simples(df["Tempo_total_UCIP_dias"], "Duração do internamento na UCIP", ylabel="Dias")

if "Ventilacao" in df.columns:
    barplot_simples(df["Ventilacao"], "Necessidade de ventilação mecânica", xlabel="Ventilação")

if "Dias de ventilação" in df.columns:
    boxplot_simples(df["Dias de ventilação"], "Duração da ventilação", ylabel="Dias")

if "Dias de sedoanalgesia" in df.columns:
    boxplot_simples(df["Dias de sedoanalgesia"], "Duração da sedoanalgesia", ylabel="Dias")

# 2.4 Avaliação pedopsiquiátrica
print("\n--- 2.4 Avaliação Pedopsiquiátrica ---")
print("Diagnóstico psiquiátrico: não existe coluna estruturada claramente identificada no dataset, por isso não será inventado.")

if "Tipo_intervencao" in df.columns:
    print("\nTipo de intervenção realizada:")
    print(tabela_frequencias(df["Tipo_intervencao"]))
    barplot_simples(df["Tipo_intervencao"], "Tipo de intervenção realizada", xlabel="Intervenção", rotate=30)
else:
    print("Tipo de intervenção: coluna não encontrada.")

# -------------------------
# 7) 3. ANÁLISE DO SEGUIMENTO PÓS-ALTA
# -------------------------
print("\n" + "="*70)
print("3. ANÁLISE DO SEGUIMENTO PÓS-ALTA")
print("="*70)

# 3.1 Continuidade de acompanhamento
print("\n--- 3.1 Continuidade de Acompanhamento ---")

if "Seguimento psic após alta" in df.columns:
    print("\nSeguimento após alta:")
    print(tabela_frequencias(df["Seguimento psic após alta"]))
    barplot_simples(df["Seguimento psic após alta"], "Seguimento após alta", xlabel="Seguimento")

if "Duração do seguimento" in df.columns:
    print("\nValores observados em 'Duração do seguimento':")
    print(df["Duração do seguimento"].value_counts(dropna=False).to_frame("n"))

print("\nTipo de seguimento: não existe coluna estruturada claramente identificada no dataset.")

# 3.2 Comparação entre doentes com e sem seguimento
print("\n--- 3.2 Comparação entre doentes com e sem seguimento ---")
print("Análise exploratória devido à dimensão amostral reduzida.")

resultados_seguimento = []

if "Seguimento psic após alta" in df.columns:
    # contínuas
    for var in ["Idade", "Tempo_total_UCIP_dias"]:
        if var in df.columns:
            res = comparar_continua_2_grupos(df, var, "Seguimento psic após alta")
            resultados_seguimento.append({
                "Variável": var,
                "Tipo": "Contínua",
                "Teste": res["teste"],
                "p": res["p"],
                "Interpretação": interpretar_p(res["p"])
            })

    # categóricas
    for var in ["Sexo", "Doença psiquiátrica prévia", "Categoria_diagnóstico", "Ventilacao"]:
        if var in df.columns:
            res = comparar_categorica_2_grupos(df, var, "Seguimento psic após alta")
            resultados_seguimento.append({
                "Variável": var,
                "Tipo": "Categórica",
                "Teste": res["teste"],
                "p": res["p"],
                "Interpretação": interpretar_p(res["p"])
            })

    resultados_seguimento_df = pd.DataFrame(resultados_seguimento)
    print("\nResultados da comparação:")
    print(resultados_seguimento_df)

    # gráficos comparativos
    if "Idade" in df.columns:
        tmp = df[["Idade", "Seguimento psic após alta"]].dropna()
        if len(tmp) > 0:
            plt.figure(figsize=(8, 5))
            sns.boxplot(data=tmp, x="Seguimento psic após alta", y="Idade")
            plt.title("Idade por seguimento após alta")
            plt.xlabel("Seguimento após alta")
            plt.ylabel("Idade")
            plt.tight_layout()
            plt.show()

    if "Tempo_total_UCIP_dias" in df.columns:
        tmp = df[["Tempo_total_UCIP_dias", "Seguimento psic após alta"]].dropna()
        if len(tmp) > 0:
            plt.figure(figsize=(8, 5))
            sns.boxplot(data=tmp, x="Seguimento psic após alta", y="Tempo_total_UCIP_dias")
            plt.title("Duração do internamento por seguimento após alta")
            plt.xlabel("Seguimento após alta")
            plt.ylabel("Dias")
            plt.tight_layout()
            plt.show()

    for var in ["Sexo", "Doença psiquiátrica prévia", "Categoria_diagnóstico", "Ventilacao"]:
        if var in df.columns:
            tmp = df[[var, "Seguimento psic após alta"]].dropna()
            if len(tmp) > 0:
                plt.figure(figsize=(10, 6))
                sns.countplot(data=tmp, x=var, hue="Seguimento psic após alta")
                plt.title(f"{var} por seguimento após alta")
                plt.xlabel(var)
                plt.ylabel("Frequência")
                plt.xticks(rotation=25, ha="right")
                plt.tight_layout()
                plt.show()

# -------------------------
# 8) 4. GRUPOS A/B/C
# -------------------------
print("\n" + "="*70)
print("4. ANÁLISE DOS GRUPOS CLÍNICOS A, B E C")
print("="*70)
print("Os grupos A/B/C não serão calculados automaticamente.")
print("Essa classificação não está explícita no dataset e não será inferida.")
print("Se quiseres analisar esta secção sem inventar nada, cria no Excel uma coluna chamada 'Grupo_clinico' com valores A, B ou C.")

# -------------------------
# 9) SUMÁRIOS FINAIS
# -------------------------
print("\n" + "="*70)
print("5. SUMÁRIOS FINAIS")
print("="*70)

resumos_lista = []

for col in ["Idade", "Tempo_total_UCIP_dias", "Dias de ventilação", "Dias de sedoanalgesia"]:
    if col in df.columns:
        resumos_lista.append(resumo_continuo(df[col], col))

if len(resumos_lista) > 0:
    resumos_continuos = pd.concat(resumos_lista, ignore_index=True)
    print("\nResumo das variáveis contínuas:")
    print(resumos_continuos)

for col in [
    "Sexo",
    "Doença psiquiátrica prévia",
    "Doença_cronica_previa_bin",
    "Categoria_diagnóstico",
    "Ventilacao",
    "Tipo_intervencao",
    "Medicação psiquiátrica iniciada",
    "Alta UCIP destino",
    "Seguimento psic após alta"
]:
    if col in df.columns:
        print(f"\n--- {col} ---")
        print(tabela_frequencias(df[col]))

# -------------------------
# 10) TEXTO METODOLÓGICO 
# -------------------------
print("\n" + "="*70)
print("TEXTO METODOLÓGICO PRONTO")
print("="*70)

texto_metodologico = """
Considerações estatísticas gerais:
- A análise foi realizada em Python.
- Foram analisados exclusivamente os 38 registos presentes na folha 'Recolha_de_Dados'.
- Não foi assumida qualquer população total externa a esta folha.
- A distribuição das variáveis contínuas foi avaliada pelo teste de Shapiro-Wilk.
- Variáveis contínuas com distribuição aproximadamente normal foram descritas por média ± desvio padrão.
- Variáveis contínuas sem distribuição normal foram descritas por mediana e intervalo interquartil.
- Na comparação entre dois grupos, foi utilizado o teste t de Student (Welch) para variáveis contínuas normais e o teste de Mann-Whitney para variáveis não normais.
- Para variáveis categóricas, foi utilizado o teste do qui-quadrado ou o teste exato de Fisher, conforme apropriado.
- Foi considerado estatisticamente significativo p < 0,05.
- Dada a reduzida dimensão amostral, as análises comparativas devem ser interpretadas com cautela e caráter exploratório.
- Não foram criadas variáveis clínicas inferidas automaticamente quando essa informação não estava explicitamente estruturada no dataset.
"""
print(texto_metodologico)
```
