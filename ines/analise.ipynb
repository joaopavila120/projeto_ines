```python
# analise_total_terminal.py
# -*- coding: utf-8 -*-

import re
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.nonparametric.smoothers_lowess import lowess


# =========================
# CONFIG
# =========================
INPUT_XLSX = "data.xlsx"
SHEET_NAME = "Folha1"

SHOW_PLOTS = True   # mantém True para aparecer na hora (VS Code / Jupyter)


# =========================
# HELPERS (sem invenção)
# =========================
def normtxt(s: str) -> str:
    s = str(s)
    s = s.strip().lower()
    # remove acentos de forma simples
    s = (s.replace("ç", "c")
           .replace("ã", "a").replace("á", "a").replace("à", "a").replace("â", "a")
           .replace("é", "e").replace("ê", "e")
           .replace("í", "i")
           .replace("ó", "o").replace("ô", "o").replace("õ", "o")
           .replace("ú", "u"))
    s = re.sub(r"\s+", " ", s)
    return s

def clean_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [re.sub(r"\s+", " ", str(c)).strip() for c in df.columns]
    return df

def find_col(df: pd.DataFrame, must_contain=None, any_of=None, exact=None):
    cols = list(df.columns)

    if exact is not None:
        for c in cols:
            if normtxt(c) == normtxt(exact):
                return c

    must_contain = must_contain or []
    any_of = any_of or []

    scored = []
    for c in cols:
        nc = normtxt(c)
        ok = True
        for token in must_contain:
            if normtxt(token) not in nc:
                ok = False
                break
        if not ok:
            continue

        score = 0
        for token in must_contain:
            score += 3
        for token in any_of:
            if normtxt(token) in nc:
                score += 1
        scored.append((score, c))

    if not scored:
        return None
    scored.sort(reverse=True, key=lambda x: x[0])
    return scored[0][1]

def to_binary_sim_nao(s: pd.Series) -> pd.Series:
    def conv(x):
        if pd.isna(x):
            return np.nan
        if isinstance(x, (int, float)) and not isinstance(x, bool):
            if x in [0, 1]:
                return int(x)
        t = normtxt(x)
        if t in ["sim", "s", "yes", "y", "1", "true"]:
            return 1
        if t in ["nao", "não", "n", "no", "0", "false"]:
            return 0
        return np.nan
    return s.apply(conv)

def as_numeric(s: pd.Series) -> pd.Series:
    return pd.to_numeric(s, errors="coerce")

def iqr(x: pd.Series) -> float:
    x = as_numeric(x).dropna()
    if len(x) == 0:
        return np.nan
    return float(x.quantile(0.75) - x.quantile(0.25))

def desc_cont(x: pd.Series):
    x = as_numeric(x).dropna()
    if len(x) == 0:
        return {"n": 0}
    out = {
        "n": int(len(x)),
        "mean": float(x.mean()),
        "sd": float(x.std(ddof=1)) if len(x) > 1 else np.nan,
        "median": float(x.median()),
        "q1": float(x.quantile(0.25)),
        "q3": float(x.quantile(0.75)),
        "iqr": float(x.quantile(0.75) - x.quantile(0.25)),
        "min": float(x.min()),
        "max": float(x.max())
    }
    return out

def shapiro_res(x: pd.Series):
    x = as_numeric(x).dropna()
    if len(x) < 3:
        return None
    W, p = stats.shapiro(x)
    return {"n": int(len(x)), "W": float(W), "p": float(p), "normal_alpha_0_05": bool(p >= 0.05)}

def print_section(title):
    print("\n" + "=" * 70)
    print(title)
    print("=" * 70)

def print_warn(msg):
    print(f"[AVISO] {msg}")

def freq_table(series: pd.Series, dropna=False):
    vc = series.value_counts(dropna=dropna)
    total = vc.sum()
    df = vc.rename_axis("categoria").reset_index(name="n")
    df["pct"] = df["n"] / total * 100 if total else np.nan
    return df

def freq_table_binary(df: pd.DataFrame, col: str):
    b = to_binary_sim_nao(df[col])
    out = b.value_counts(dropna=False).rename_axis("value").reset_index(name="n")
    total = out["n"].sum()
    out["pct"] = out["n"] / total * 100 if total else np.nan
    return out

def mann_whitney_clean(a: pd.Series, b: pd.Series):
    a = as_numeric(a).dropna()
    b = as_numeric(b).dropna()
    if len(a) == 0 or len(b) == 0:
        return None
    U, p = stats.mannwhitneyu(a, b, alternative="two-sided", method="auto")
    return {
        "n1": int(len(a)),
        "n2": int(len(b)),
        "mediana1": float(a.median()),
        "mediana2": float(b.median()),
        "U": float(U),
        "p": float(p)
    }

def spearman_clean(x: pd.Series, y: pd.Series):
    x = as_numeric(x)
    y = as_numeric(y)
    m = (~x.isna()) & (~y.isna())
    if m.sum() < 3:
        return None
    rho, p = stats.spearmanr(x[m], y[m])
    return {"n": int(m.sum()), "rho": float(rho), "p": float(p)}

def pearson_clean(x: pd.Series, y: pd.Series):
    x = as_numeric(x)
    y = as_numeric(y)
    m = (~x.isna()) & (~y.isna())
    if m.sum() < 3:
        return None
    r, p = stats.pearsonr(x[m], y[m])
    return {"n": int(m.sum()), "r": float(r), "p": float(p)}

def kruskal_by_group(df: pd.DataFrame, group_col: str, value_col: str):
    tmp = df[[group_col, value_col]].copy()
    tmp[value_col] = as_numeric(tmp[value_col])
    tmp = tmp.dropna()
    if tmp[group_col].nunique() < 2:
        return None
    groups = [g[value_col].values for _, g in tmp.groupby(group_col)]
    if any(len(arr) == 0 for arr in groups):
        return None
    H, p = stats.kruskal(*groups)
    return {"k": int(tmp[group_col].nunique()), "H": float(H), "p": float(p), "n": int(len(tmp))}

def holm_adjust(pvals):
    # Holm-Bonferroni
    pvals = np.array(pvals, dtype=float)
    m = len(pvals)
    order = np.argsort(pvals)
    adj = np.empty(m, dtype=float)
    for i, idx in enumerate(order):
        adj[idx] = min(1.0, (m - i) * pvals[idx])
    # monotonicidade
    for i in range(m - 2, -1, -1):
        adj[order[i]] = max(adj[order[i]], adj[order[i + 1]])
    return adj

def posthoc_pairwise_mw(df: pd.DataFrame, group_col: str, value_col: str):
    # Pós-hoc simples: Mann-Whitney pairwise + Holm
    tmp = df[[group_col, value_col]].copy()
    tmp[value_col] = as_numeric(tmp[value_col])
    tmp = tmp.dropna()
    groups = list(tmp[group_col].unique())
    if len(groups) < 3:
        return None

    pairs = []
    pvals = []
    for i in range(len(groups)):
        for j in range(i + 1, len(groups)):
            g1, g2 = groups[i], groups[j]
            a = tmp.loc[tmp[group_col] == g1, value_col]
            b = tmp.loc[tmp[group_col] == g2, value_col]
            res = mann_whitney_clean(a, b)
            if res is None:
                continue
            pairs.append((g1, g2, res["p"]))
            pvals.append(res["p"])

    if not pairs:
        return None

    adj = holm_adjust(pvals)
    out = []
    for (g1, g2, p), padj in zip(pairs, adj):
        out.append({"grupo1": g1, "grupo2": g2, "p_raw": float(p), "p_holm": float(padj)})
    return pd.DataFrame(out).sort_values("p_holm")

def chi2_or_fisher_2x2(a: pd.Series, b: pd.Series):
    a = to_binary_sim_nao(a)
    b = to_binary_sim_nao(b)
    tab = pd.crosstab(a, b).reindex(index=[0, 1], columns=[0, 1], fill_value=0)
    # chi2 expected
    chi2, p_chi2, dof, expected = stats.chi2_contingency(tab.values)
    expected_min = expected.min() if expected.size else np.nan
    res = {
        "t00": int(tab.loc[0, 0]), "t01": int(tab.loc[0, 1]),
        "t10": int(tab.loc[1, 0]), "t11": int(tab.loc[1, 1]),
        "chi2": float(chi2), "p_chi2": float(p_chi2), "expected_min": float(expected_min),
        "teste_usado": "chi2", "p_usado": float(p_chi2)
    }
    if expected_min < 5:
        odds, p_f = stats.fisher_exact(tab.values)
        res["teste_usado"] = "fisher"
        res["oddsratio"] = float(odds)
        res["p_usado"] = float(p_f)
    return res

def plot_hist_box(x: pd.Series, title: str):
    x = as_numeric(x).dropna()
    if len(x) == 0:
        return
    plt.figure()
    plt.hist(x, bins="auto")
    plt.title(f"Histograma: {title}")
    plt.xlabel(title)
    plt.ylabel("Frequência")
    if SHOW_PLOTS:
        plt.show()
    plt.close()

    plt.figure()
    plt.boxplot(x, vert=True)
    plt.title(f"Boxplot: {title}")
    plt.ylabel(title)
    if SHOW_PLOTS:
        plt.show()
    plt.close()

def plot_bar_counts(labels, counts, title, xlabel):
    plt.figure()
    plt.bar(labels, counts)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("n")
    if SHOW_PLOTS:
        plt.show()
    plt.close()

def plot_scatter_trend(x: pd.Series, y: pd.Series, title: str, xlabel: str, ylabel: str):
    x = as_numeric(x)
    y = as_numeric(y)
    m = (~x.isna()) & (~y.isna())
    x = x[m].values
    y = y[m].values
    if len(x) < 3:
        return

    plt.figure()
    plt.scatter(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # tendência LOWESS (boa mesmo se não-linear)
    try:
        smoothed = lowess(y, x, frac=0.6, return_sorted=True)
        plt.plot(smoothed[:, 0], smoothed[:, 1])
    except Exception:
        # fallback: reta
        m1, b1 = np.polyfit(x, y, 1)
        xs = np.linspace(x.min(), x.max(), 100)
        plt.plot(xs, m1 * xs + b1)

    if SHOW_PLOTS:
        plt.show()
    plt.close()

def is_normal_by_shapiro(x: pd.Series, alpha=0.05):
    r = shapiro_res(x)
    if r is None:
        return None
    return bool(r["p"] >= alpha)


# =========================
# LOAD
# =========================
df = pd.read_excel(INPUT_XLSX, sheet_name=SHEET_NAME)
df = clean_columns(df)

print_section("COLUNAS ENCONTRADAS NO EXCEL (para transparência)")
print(pd.Series(df.columns).to_string(index=False))


# =========================
# AUTO-MAPEAMENTO DE COLUNAS (sem inventar)
# =========================
col_sexo = find_col(df, exact="Sexo") or find_col(df, must_contain=["sexo"])
col_idade = find_col(df, exact="Idade") or find_col(df, must_contain=["idade"])
col_nat = find_col(df, exact="naturalidade") or find_col(df, must_contain=["naturalidade"])
col_zona = find_col(df, must_contain=["zona"], any_of=["residencia", "residência"])

col_intern_ano = find_col(df, must_contain=["media", "internamentos", "por ano"])
col_su_ano = find_col(df, must_contain=["media", "idas", "su", "por ano"]) or find_col(df, must_contain=["media", "su", "por ano"])

col_hb = find_col(df, must_contain=["hb", "basal"]) or find_col(df, exact="Hb basal")
col_hbs = find_col(df, must_contain=["hbs", "basal"]) or find_col(df, exact="HbS basal")

col_cvo_media_ano = find_col(df, must_contain=["media", "cvo", "por ano"])
col_intern_por_cvo_ano = find_col(df, must_contain=["media", "internamentos", "por", "cvo", "por ano"])
col_cvo_total = find_col(df, must_contain=["cvo"], any_of=["nº total", "no total", "total", "n total", "numero total", "número total"])

col_avc_total = find_col(df, must_contain=["avc"], any_of=["nº total", "total", "n total", "numero total", "número total"]) or find_col(df, exact="AVC (nº total)")
col_avc_tipo = find_col(df, must_contain=["avc"], any_of=["tipo", "isquem", "hemorr"])

col_inf_total = find_col(df, must_contain=["infe"], any_of=["bacter", "nº total", "total"]) or find_col(df, exact="Infeções bacterianas (nº total)")
col_inf_sistema = find_col(df, must_contain=["infe"], any_of=["sistema", "afetado", "local", "orgao", "órgão", "tipo"])

col_anemia_total = find_col(df, must_contain=["anemia"], any_of=["agud", "nº total", "total"])
col_anemia_intern = find_col(df, must_contain=["intern"], any_of=["anemia"])

col_sta_total = find_col(df, must_contain=["sta"], any_of=["nº total", "total", "nºtotal"])
col_uci_total = find_col(df, must_contain=["uci"], any_of=["sta", "necessidade", "total"])

# Vacinas (tenta achar por exato; se não achar, tenta aproximar)
vacinas_target = ["Pn", "Hib", "Men B", "Men ACWY", "Gripe", "SARS-CoV-2", "SARS-COV2"]
vaccine_cols = {}
for v in vacinas_target:
    c = find_col(df, exact=v)
    if c is None:
        # fallback por tokens
        tokens = [t for t in re.split(r"\s+|-|_", v) if t]
        c = find_col(df, must_contain=tokens)
    if c is not None:
        vaccine_cols[v] = c

# Hidroxiureia e dose
col_hu = find_col(df, must_contain=["hidroxi"]) or find_col(df, exact="Hidroxiureia")
col_hu_dose = find_col(df, must_contain=["dos"], any_of=["mg", "dia", "hidroxi"]) or find_col(df, exact="Dosagem (mg/dia)")

# Transfusões e tipo/frequência
col_transf = find_col(df, must_contain=["transfus"])
col_transf_tipo = find_col(df, must_contain=["transfus"], any_of=["tipo", "frequ", "ocas", "mens"])

# Crónicas: lista (usa match por normalização do texto)
chronic_names = [
    "DRC",
    "Úlcera Cutânea Crónica",
    "Hepatopatia falciforme",
    "Hepatosiderose",
    "Litíase Vesicular",
    "Doença Pulmonar Obstrutiva",
    "Doença pulmonar restritiva",
    "Osteopenia ou osteoporose",
    "Necrose avascular óssea",
    "Osteomielite crónica",
    "Cardiopatia dilatada",
    "Cardiopatia hipertrófica",
    "Doença cerebrovascular",
    "Hipertensão pulmonar",
    "Retinopatia ou máculopatia",
    "Disfunção Erétil",
]
chronic_cols = {}
for name in chronic_names:
    c = find_col(df, exact=name)
    if c is None:
        tokens = [t for t in normtxt(name).split(" ") if len(t) > 2]
        c = find_col(df, must_contain=tokens[:2], any_of=tokens[2:]) if tokens else None
    if c is not None:
        chronic_cols[name] = c

# Osteomielite vs artroplastia
col_osteom = chronic_cols.get("Osteomielite crónica") or find_col(df, must_contain=["osteomiel"], any_of=["cron", "cronic"])
col_artro = find_col(df, must_contain=["artropl"]) or find_col(df, exact="Artroplastia (SIM/NÃO)")

# Ferro hepático
col_ferro = find_col(df, must_contain=["ferro", "hep"]) or find_col(df, must_contain=["ferro"], any_of=["hepatico", "hepática"])

# Rastreios
screening_targets = [
    "RM crânio_pescoço",
    "ECOTT",
    "Fundoscoppia",
    "RM Hepática",
    "ECO Renal",
    "Provas Respiratórias",
    "Densitometria",
    "RM crânio e pescoço",
    "ECO TT",
    "Fundoscopia",
]
screening_cols = {}
for sname in screening_targets:
    c = find_col(df, exact=sname)
    if c is None:
        tokens = [t for t in normtxt(sname).replace("_", " ").split(" ") if len(t) > 1]
        c = find_col(df, must_contain=tokens[:2], any_of=tokens[2:]) if tokens else None
    if c is not None:
        screening_cols[sname] = c

# Marcadores de gravidade (para associações)
col_intern_por_cvo = col_intern_por_cvo_ano  # usado como marcador
col_cvo_marker = col_cvo_total or col_cvo_media_ano  # preferir total se existir


# =========================
# 0) NORMALIDADE DE TODAS AS CONTÍNUAS (SHAPIRO)
# =========================
print_section("0) NORMALIDADE (Shapiro-Wilk) - variáveis contínuas detectadas")
# heurística: colunas numéricas ou que viram numéricas com muitos valores
cont_candidates = []
for c in df.columns:
    s = as_numeric(df[c])
    n_num = int(s.notna().sum())
    if n_num >= max(10, int(0.3 * len(df))):  # pelo menos 30% ou 10 valores
        cont_candidates.append(c)

if not cont_candidates:
    print_warn("Nenhuma variável contínua detectada automaticamente com confiança.")
else:
    for c in cont_candidates:
        r = shapiro_res(df[c])
        if r is None:
            print(f"- {c}: Shapiro não aplicável (n<3 numérico)")
        else:
            print(f"- {c}: n={r['n']} W={r['W']:.4f} p={r['p']:.6g} => normal? {r['normal_alpha_0_05']}")


# =========================
# 1) DESCRIÇÃO DA POPULAÇÃO
# =========================
print_section("1) DESCRIÇÃO DA POPULAÇÃO")

if col_sexo:
    print("\nSEXO (n e %)")
    print(freq_table(df[col_sexo], dropna=False).to_string(index=False))
else:
    print_warn("Coluna de SEXO não encontrada.")

if col_idade:
    print("\nIDADE: média±DP e mediana")
    d = desc_cont(df[col_idade])
    print(f"n={d.get('n',0)}  média={d.get('mean',np.nan):.3f}  DP={d.get('sd',np.nan):.3f}  mediana={d.get('median',np.nan):.3f}")
    print("Shapiro:", shapiro_res(df[col_idade]))
else:
    print_warn("Coluna de IDADE não encontrada.")

if col_nat:
    print("\nNATURALIDADE (n e %)")
    print(freq_table(df[col_nat], dropna=False).to_string(index=False))
else:
    print_warn("Coluna de NATURALIDADE não encontrada.")

if col_zona:
    print("\nZONA DE RESIDÊNCIA (n e %)")
    print(freq_table(df[col_zona], dropna=False).to_string(index=False))
else:
    print_warn("Coluna de ZONA DE RESIDÊNCIA não encontrada.")

def print_med_iqr_minmax(col, label):
    if not col:
        print_warn(f"{label}: coluna não encontrada.")
        return
    d = desc_cont(df[col])
    print(f"\n{label}: mediana (IQR), min–max  [esperado assimétrico]")
    print(f"n={d.get('n',0)}  mediana={d.get('median',np.nan):.3f}  IQR={d.get('iqr',np.nan):.3f}  min={d.get('min',np.nan):.3f}  max={d.get('max',np.nan):.3f}")
    print("Shapiro:", shapiro_res(df[col]))

print_med_iqr_minmax(col_intern_ano, "Média de internamentos por ano")
print_med_iqr_minmax(col_su_ano, "Média de idas ao SU por ano")
print_med_iqr_minmax(col_hb, "Hb basal")
print_med_iqr_minmax(col_hbs, "HbS basal")


# =========================
# 2) MANIFESTAÇÕES AGUDAS
# =========================
print_section("2) MANIFESTAÇÕES AGUDAS")

# 2.1 CVO/ano: descritivo + Shapiro + hist/box
if col_cvo_media_ano:
    print("\nMédia de CVO por ano")
    d = desc_cont(df[col_cvo_media_ano])
    sh = shapiro_res(df[col_cvo_media_ano])
    print("Descritivo:", d)
    print("Shapiro:", sh)
    plot_hist_box(df[col_cvo_media_ano], "Média de CVO por ano")
else:
    print_warn("Coluna 'média de CVO por ano' não encontrada.")

# 2.2 Internamentos por CVO/ano + proporção CVO com internamento
if col_intern_por_cvo_ano:
    print("\nMédia de internamentos por CVO por ano")
    d = desc_cont(df[col_intern_por_cvo_ano])
    print("Descritivo:", d)
    print("Shapiro:", shapiro_res(df[col_intern_por_cvo_ano]))
    if col_cvo_media_ano:
        # proporção aproximada = internamentos_CVO_ano / CVO_ano (se ambos existirem)
        a = as_numeric(df[col_intern_por_cvo_ano])
        b = as_numeric(df[col_cvo_media_ano])
        m = (~a.isna()) & (~b.isna()) & (b > 0)
        if m.sum() > 0:
            prop = (a[m] / b[m])
            print("\nProporção estimada de CVO que resultaram em internamento (internamentos/CVO):")
            print(desc_cont(prop))
            # barras (com internamento vs sem) a partir da proporção média
            # aqui: só dá para mostrar distribuição da proporção, não contagem real.
            plot_hist_box(prop, "Proporção internamentos/CVO (estimada)")
        else:
            print_warn("Não foi possível calcular proporção internamentos/CVO (faltam dados ou CVO=0).")
else:
    print_warn("Coluna 'média de internamentos por CVO por ano' não encontrada.")

# 2.3 AVC total + proporções 0/≥1/≥2… + tipo
if col_avc_total:
    avc = as_numeric(df[col_avc_total])
    print("\nNº total de AVC por doente")
    print("Descritivo:", desc_cont(avc))
    print("Shapiro:", shapiro_res(avc))

    # Proporções 0, >=1, >=2, >=3...
    avc_int = avc.dropna()
    if len(avc_int) > 0:
        thresholds = [0, 1, 2, 3]
        print("\nProporção de doentes por categorias de AVC:")
        rows = []
        for t in thresholds:
            if t == 0:
                n = int((avc_int == 0).sum())
                lab = "0"
            else:
                n = int((avc_int >= t).sum())
                lab = f">={t}"
            rows.append({"categoria": lab, "n": n, "pct": n / len(avc_int) * 100})
        print(pd.DataFrame(rows).to_string(index=False))
else:
    print_warn("Coluna de 'Nº total AVC' não encontrada.")

if col_avc_tipo:
    print("\nTipo de AVC (frequência)")
    print(freq_table(df[col_avc_tipo], dropna=False).to_string(index=False))
else:
    print_warn("Coluna de 'tipo de AVC' não encontrada.")

# 2.4 Infeções totais + sistema afetado
if col_inf_total:
    inf = as_numeric(df[col_inf_total])
    print("\nNº total de infeções bacterianas")
    print("Descritivo:", desc_cont(inf))
    print("Shapiro:", shapiro_res(inf))
else:
    print_warn("Coluna de infeções totais não encontrada.")

if col_inf_sistema:
    print("\nSistema afetado nas infeções (frequência)")
    print(freq_table(df[col_inf_sistema], dropna=False).to_string(index=False))
else:
    print_warn("Coluna de 'sistema afetado' das infeções não encontrada.")

# 2.5 Vacinação (n e %) + Vacinação vs infeções (Mann-Whitney)
print_section("2.5) VACINAÇÃO (n e %) + Vacinação vs Infeções (Mann-Whitney; exploratório, sem causalidade)")

if not vaccine_cols:
    print_warn("Nenhuma coluna de vacina encontrada pelos nomes/padrões comuns.")
else:
    if not col_inf_total:
        print_warn("Sem coluna de infeções totais => não dá para fazer Vacinação vs Infeções.")
    for vname, vcol in vaccine_cols.items():
        print(f"\nVacina: {vname}  [coluna: {vcol}]")
        tab = freq_table_binary(df, vcol)
        print(tab.to_string(index=False))
        if col_inf_total:
            b = to_binary_sim_nao(df[vcol])
            res = mann_whitney_clean(df.loc[b == 1, col_inf_total], df.loc[b == 0, col_inf_total])
            if res is None:
                print_warn("Mann-Whitney não foi possível (um grupo sem dados numéricos suficientes).")
            else:
                print("Mann-Whitney (nº infeções em vacinados vs não):", res)

# 2.6 Anemia agudizada + internamentos por anemia
print_section("2.6) ANEMIA AGUDIZADA / INTERNAMENTOS POR ANEMIA")

if col_anemia_total:
    print("\nNº total de agudizações de anemia")
    print("Descritivo:", desc_cont(df[col_anemia_total]))
    print("Shapiro:", shapiro_res(df[col_anemia_total]))
else:
    print_warn("Coluna 'nº total de agudizações de anemia' não encontrada.")

if col_anemia_intern:
    print("\nNº total de internamentos por anemia agudizada")
    print("Descritivo:", desc_cont(df[col_anemia_intern]))
    print("Shapiro:", shapiro_res(df[col_anemia_intern]))
else:
    print_warn("Coluna 'internamentos por anemia agudizada' não encontrada.")

# 2.7 STA + UCI
print_section("2.7) STA / NECESSIDADE DE UCI")

if col_sta_total:
    print("\nNº total de STA")
    print("Descritivo:", desc_cont(df[col_sta_total]))
    print("Shapiro:", shapiro_res(df[col_sta_total]))
else:
    print_warn("Coluna 'STA total' não encontrada.")

if col_uci_total:
    print("\nNº total de necessidade de UCI (relacionada a STA, se aplicável)")
    print("Descritivo:", desc_cont(df[col_uci_total]))
    print("Shapiro:", shapiro_res(df[col_uci_total]))
else:
    print_warn("Coluna 'necessidade de UCI' não encontrada.")


# =========================
# 3) Hb / HbS como marcadores de gravidade (correlação + regressão + gráficos)
# =========================
print_section("3) Hb basal / HbS basal como marcadores (correlação + regressão Poisson/NegBin + gráficos)")

if not col_cvo_marker:
    print_warn("Não encontrei variável de CVO (total ou média/ano).")
else:
    ycol = col_cvo_marker
    print(f"Variável de CVO usada como outcome: {ycol}")

    # correlação Hb vs CVO e HbS vs CVO (Pearson se ambos normais, senão Spearman)
    def corr_choose(xcol, ycol, xname):
        if not xcol:
            print_warn(f"{xname}: coluna não encontrada.")
            return

        x = df[xcol]
        y = df[ycol]
        nx = is_normal_by_shapiro(x)
        ny = is_normal_by_shapiro(y)

        if (nx is True) and (ny is True):
            r = pearson_clean(x, y)
            print(f"\nCorrelação Pearson: {xname} ↔ CVO")
            print(r if r else "[sem dados suficientes]")
        else:
            r = spearman_clean(x, y)
            print(f"\nCorrelação Spearman: {xname} ↔ CVO (não-normal ou indeterminado)")
            print(r if r else "[sem dados suficientes]")

        plot_scatter_trend(x, y,
                           title=f"{xname} vs CVO",
                           xlabel=xname,
                           ylabel="CVO")

    corr_choose(col_hb, ycol, "Hb basal")
    corr_choose(col_hbs, ycol, "HbS basal")

    # Regressão: Poisson e NegBin (ideal para contagem)
    # Se a coluna de CVO não for contagem inteira, avisa.
    y = as_numeric(df[ycol])
    if y.dropna().shape[0] > 0:
        # checa se parece contagem
        frac_nonint = float(((y.dropna() % 1) != 0).mean()) if len(y.dropna()) else 0.0
        if frac_nonint > 0.2:
            print_warn("Outcome de CVO tem muitos valores não-inteiros; Poisson/NegBin ainda rodam, mas interpretação como contagem fica limitada.")

    # monta dataframe de regressão
    reg_cols = []
    if col_hb: reg_cols.append(col_hb)
    if col_hbs: reg_cols.append(col_hbs)
    reg_cols.append(ycol)

    reg_df = df[reg_cols].copy()
    for c in reg_cols:
        reg_df[c] = as_numeric(reg_df[c])
    reg_df = reg_df.dropna()

    if len(reg_df) < 10:
        print_warn("Poucos dados para regressão (n<10 após limpar NaNs).")
    else:
        # renomeia para fórmula
        tmp = reg_df.rename(columns={ycol: "cvo"})
        if col_hb: tmp = tmp.rename(columns={col_hb: "hb"})
        if col_hbs: tmp = tmp.rename(columns={col_hbs: "hbs"})

        rhs = []
        if "hb" in tmp.columns: rhs.append("hb")
        if "hbs" in tmp.columns: rhs.append("hbs")
        formula = "cvo ~ " + " + ".join(rhs) if rhs else "cvo ~ 1"

        print("\nRegressão Poisson (GLM):", formula)
        try:
            pois = smf.glm(formula=formula, data=tmp, family=sm.families.Poisson()).fit()
            print(pois.summary())
        except Exception as e:
            print_warn(f"Poisson falhou: {e}")

        print("\nRegressão Binomial Negativa (GLM):", formula)
        try:
            nb = smf.glm(formula=formula, data=tmp, family=sm.families.NegativeBinomial()).fit()
            print(nb.summary())
        except Exception as e:
            print_warn(f"NegBin falhou: {e}")


# =========================
# 4) MANIFESTAÇÕES CRÓNICAS: tabela única + associações + osteomielite vs artroplastia + ferro hepático
# =========================
print_section("4) MANIFESTAÇÕES CRÓNICAS: prevalências (Tabela 2) + associações + osteomielite/artroplastia + ferro hepático")

if not chronic_cols:
    print_warn("Não foi possível mapear as crónicas (nenhuma coluna correspondente encontrada).")
else:
    # Tabela 2 (n e % SIM)
    rows = []
    for label, col in chronic_cols.items():
        b = to_binary_sim_nao(df[col])
        n_total = int(b.notna().sum())
        n_sim = int((b == 1).sum())
        pct = (n_sim / n_total * 100) if n_total else np.nan
        rows.append({"manifestacao": label, "coluna": col, "n_total_com_dado": n_total, "n_SIM": n_sim, "pct_SIM": pct})
    table2 = pd.DataFrame(rows).sort_values("pct_SIM", ascending=False)

    print("\nTabela 2 - Crónicas (n e %):")
    print(table2.to_string(index=False))

    # Gráfico opcional: barras horizontais das mais frequentes
    top = table2.head(10).copy()
    if len(top) > 0:
        plt.figure()
        plt.barh(top["manifestacao"][::-1], top["pct_SIM"][::-1])
        plt.title("Top 10 manifestações crónicas (% SIM)")
        plt.xlabel("% SIM")
        if SHOW_PLOTS:
            plt.show()
        plt.close()

    # Burden = soma das crónicas (NaN tratado como 0)
    chronic_bin = pd.DataFrame()
    for label, col in chronic_cols.items():
        chronic_bin[label] = to_binary_sim_nao(df[col]).fillna(0).astype(int)
    df["burden_cronico"] = chronic_bin.sum(axis=1)

    print("\nBurden crónico (soma das crónicas):")
    print(desc_cont(df["burden_cronico"]))
    print("Distribuição burden:")
    print(freq_table(df["burden_cronico"], dropna=False).to_string(index=False))

    # Associação com marcadores (Spearman) e scatter
    markers = []
    if col_cvo_marker: markers.append(("CVO", col_cvo_marker))
    if col_intern_por_cvo: markers.append(("Internamentos por CVO/ano", col_intern_por_cvo))
    if col_hb: markers.append(("Hb basal", col_hb))
    if col_hbs: markers.append(("HbS basal", col_hbs))

    if not markers:
        print_warn("Sem marcadores suficientes para associação com burden.")
    else:
        print("\nSpearman: burden ↔ marcadores")
        for mlabel, mcol in markers:
            r = spearman_clean(df["burden_cronico"], df[mcol])
            print(f"- burden vs {mlabel} [{mcol}]:", r if r else "[sem dados suficientes]")
            plot_scatter_trend(df["burden_cronico"], df[mcol],
                               title=f"burden vs {mlabel}",
                               xlabel="burden_cronico",
                               ylabel=mlabel)

    # Osteomielite vs Artroplastia (qui2/fisher)
    if col_osteom and col_artro:
        print("\nOsteomielite crónica vs Artroplastia (Qui² ou Fisher):")
        res = chi2_or_fisher_2x2(df[col_osteom], df[col_artro])
        print(res)
    else:
        print_warn("Teste osteomielite vs artroplastia: faltou coluna de osteomielite ou de artroplastia.")

    # Ferro hepático apenas nos com hepatosiderose (descritivo)
    col_hepatosid = chronic_cols.get("Hepatosiderose")
    if col_hepatosid and col_ferro:
        b = to_binary_sim_nao(df[col_hepatosid])
        ferro = as_numeric(df[col_ferro])
        subset = ferro[b == 1]
        print("\nFerro hepático (somente em doentes com hepatosiderose):")
        print(desc_cont(subset))
        plot_hist_box(subset, "Ferro hepático (em hepatosiderose)")
    else:
        print_warn("Ferro hepático em hepatosiderose: faltou coluna Hepatosiderose e/ou Ferro hepático.")


# =========================
# 5) BURDEN por grupos + Kruskal-Wallis + pós-hoc
# =========================
print_section("5) BURDEN por grupos (0, 1-2, >=3) + Kruskal-Wallis + pós-hoc (Holm)")

if "burden_cronico" not in df.columns:
    print_warn("Burden não foi calculado (faltam crónicas).")
else:
    def burden_group(x):
        if pd.isna(x):
            return np.nan
        x = int(x)
        if x == 0:
            return "0"
        if 1 <= x <= 2:
            return "1-2"
        return ">=3"

    df["burden_grupo"] = df["burden_cronico"].apply(burden_group)

    print("Distribuição burden_grupo:")
    print(freq_table(df["burden_grupo"], dropna=False).to_string(index=False))

    # Kruskal-Wallis para marcadores
    if col_cvo_marker:
        print("\nKruskal-Wallis: CVO por burden_grupo")
        kw = kruskal_by_group(df, "burden_grupo", col_cvo_marker)
        print(kw if kw else "[não aplicável]")
        if kw and kw["p"] < 0.05:
            ph = posthoc_pairwise_mw(df, "burden_grupo", col_cvo_marker)
            if ph is not None:
                print("Pós-hoc (Mann-Whitney pairwise, Holm):")
                print(ph.to_string(index=False))
        # boxplot
        tmp = df[["burden_grupo", col_cvo_marker]].copy()
        tmp[col_cvo_marker] = as_numeric(tmp[col_cvo_marker])
        tmp = tmp.dropna()
        if tmp["burden_grupo"].nunique() >= 2:
            plt.figure()
            groups = ["0", "1-2", ">=3"]
            data = [tmp.loc[tmp["burden_grupo"] == g, col_cvo_marker].values for g in groups if g in tmp["burden_grupo"].unique()]
            labels = [g for g in groups if g in tmp["burden_grupo"].unique()]
            plt.boxplot(data, labels=labels)
            plt.title("CVO por burden_grupo")
            plt.xlabel("burden_grupo")
            plt.ylabel("CVO")
            if SHOW_PLOTS:
                plt.show()
            plt.close()

    if col_intern_por_cvo:
        print("\nKruskal-Wallis: Internamentos por CVO/ano por burden_grupo")
        kw = kruskal_by_group(df, "burden_grupo", col_intern_por_cvo)
        print(kw if kw else "[não aplicável]")
        if kw and kw["p"] < 0.05:
            ph = posthoc_pairwise_mw(df, "burden_grupo", col_intern_por_cvo)
            if ph is not None:
                print("Pós-hoc (Mann-Whitney pairwise, Holm):")
                print(ph.to_string(index=False))

    if col_hb:
        print("\nKruskal-Wallis: Hb basal por burden_grupo")
        kw = kruskal_by_group(df, "burden_grupo", col_hb)
        print(kw if kw else "[não aplicável]")
        if kw and kw["p"] < 0.05:
            ph = posthoc_pairwise_mw(df, "burden_grupo", col_hb)
            if ph is not None:
                print("Pós-hoc (Mann-Whitney pairwise, Holm):")
                print(ph.to_string(index=False))

    if col_hbs:
        print("\nKruskal-Wallis: HbS basal por burden_grupo")
        kw = kruskal_by_group(df, "burden_grupo", col_hbs)
        print(kw if kw else "[não aplicável]")
        if kw and kw["p"] < 0.05:
            ph = posthoc_pairwise_mw(df, "burden_grupo", col_hbs)
            if ph is not None:
                print("Pós-hoc (Mann-Whitney pairwise, Holm):")
                print(ph.to_string(index=False))


# =========================
# 6) HIDROXIUREIA
# =========================
print_section("6) TERAPÊUTICA - HIDROXIUREIA")

if not col_hu:
    print_warn("Coluna Hidroxiureia (SIM/NÃO) não encontrada.")
else:
    print(f"Hidroxiureia (coluna: {col_hu}) - frequência (n e %)")
    print(freq_table_binary(df, col_hu).to_string(index=False))

    b_hu = to_binary_sim_nao(df[col_hu])

    if not col_hu_dose:
        print_warn("Coluna de dose de hidroxiureia não encontrada.")
    else:
        dose = as_numeric(df[col_hu_dose])
        dose_treated = dose[b_hu == 1]
        print("\nDistribuição da dose (somente tratados):")
        print(desc_cont(dose_treated))

        # Spearman: dose vs CVO/ano; dose vs internamentos/CVO; dose vs burden
        if col_cvo_media_ano:
            r = spearman_clean(dose_treated, df.loc[b_hu == 1, col_cvo_media_ano])
            print("\nSpearman: dose HU vs média de CVO/ano (tratados):", r if r else "[sem dados suficientes]")
            plot_scatter_trend(dose_treated, df.loc[b_hu == 1, col_cvo_media_ano],
                               title="Dose HU vs média de CVO/ano (tratados)",
                               xlabel="Dose HU (mg/dia)",
                               ylabel="média CVO/ano")

        if col_intern_por_cvo_ano:
            r = spearman_clean(dose_treated, df.loc[b_hu == 1, col_intern_por_cvo_ano])
            print("\nSpearman: dose HU vs média internamentos por CVO/ano (tratados):", r if r else "[sem dados suficientes]")
            plot_scatter_trend(dose_treated, df.loc[b_hu == 1, col_intern_por_cvo_ano],
                               title="Dose HU vs internamentos por CVO/ano (tratados)",
                               xlabel="Dose HU (mg/dia)",
                               ylabel="internamentos por CVO/ano")

        if "burden_cronico" in df.columns:
            r = spearman_clean(dose_treated, df.loc[b_hu == 1, "burden_cronico"])
            print("\nSpearman: dose HU vs burden (tratados):", r if r else "[sem dados suficientes]")
            plot_scatter_trend(dose_treated, df.loc[b_hu == 1, "burden_cronico"],
                               title="Dose HU vs burden crónico (tratados)",
                               xlabel="Dose HU (mg/dia)",
                               ylabel="burden_cronico")

            # boxplot da dose por burden_grupo (tratados)
            if "burden_grupo" in df.columns:
                tmp = df.loc[b_hu == 1, ["burden_grupo", col_hu_dose]].copy()
                tmp[col_hu_dose] = as_numeric(tmp[col_hu_dose])
                tmp = tmp.dropna()
                if tmp["burden_grupo"].nunique() >= 2:
                    plt.figure()
                    groups = ["0", "1-2", ">=3"]
                    data = [tmp.loc[tmp["burden_grupo"] == g, col_hu_dose].values for g in groups if g in tmp["burden_grupo"].unique()]
                    labels = [g for g in groups if g in tmp["burden_grupo"].unique()]
                    plt.boxplot(data, labels=labels)
                    plt.title("Dose HU por burden_grupo (tratados)")
                    plt.xlabel("burden_grupo")
                    plt.ylabel("Dose HU (mg/dia)")
                    if SHOW_PLOTS:
                        plt.show()
                    plt.close()


# =========================
# 7) TRANSFUSÕES
# =========================
print_section("7) TRANSFUSÕES")

if not col_transf:
    print_warn("Coluna de Transfusões (SIM/NÃO) não encontrada.")
else:
    print(f"Transfusões (coluna: {col_transf}) - frequência (n e %)")
    print(freq_table_binary(df, col_transf).to_string(index=False))

    b_tr = to_binary_sim_nao(df[col_transf])

    if not col_transf_tipo:
        print_warn("Coluna de tipo/frequência de transfusão não encontrada (ocasional/mensal).")
    else:
        print("\nTipo/frequência de transfusão (somente transfundidos):")
        sub = df.loc[b_tr == 1, col_transf_tipo]
        print(freq_table(sub, dropna=False).to_string(index=False))

        # Associação tipo vs CVO/ano, internamentos/CVO e burden
        outcomes = []
        if col_cvo_media_ano: outcomes.append(("média CVO/ano", col_cvo_media_ano))
        if col_intern_ano: outcomes.append(("internamentos/ano", col_intern_ano))
        if "burden_cronico" in df.columns: outcomes.append(("burden_cronico", "burden_cronico"))

        for oname, ocol in outcomes:
            tmp = df.loc[b_tr == 1, [col_transf_tipo, ocol]].copy()
            tmp[ocol] = as_numeric(tmp[ocol])
            tmp = tmp.dropna()
            if tmp[col_transf_tipo].nunique() < 2:
                print_warn(f"Sem variação de tipo de transfusão para testar vs {oname}.")
                continue

            # se 2 grupos -> Mann-Whitney; se >2 -> Kruskal
            if tmp[col_transf_tipo].nunique() == 2:
                gvals = list(tmp[col_transf_tipo].unique())
                a = tmp.loc[tmp[col_transf_tipo] == gvals[0], ocol]
                b = tmp.loc[tmp[col_transf_tipo] == gvals[1], ocol]
                res = mann_whitney_clean(a, b)
                print(f"\nTipo transfusão vs {oname} (Mann-Whitney): grupos={gvals}")
                print(res if res else "[não aplicável]")
            else:
                kw = kruskal_by_group(tmp, col_transf_tipo, ocol)
                print(f"\nTipo transfusão vs {oname} (Kruskal-Wallis):")
                print(kw if kw else "[não aplicável]")
                if kw and kw["p"] < 0.05:
                    ph = posthoc_pairwise_mw(tmp, col_transf_tipo, ocol)
                    if ph is not None:
                        print("Pós-hoc (Mann-Whitney pairwise, Holm):")
                        print(ph.to_string(index=False))

            # boxplot
            plt.figure()
            cats = list(tmp[col_transf_tipo].unique())
            data = [tmp.loc[tmp[col_transf_tipo] == c, ocol].values for c in cats]
            plt.boxplot(data, labels=cats)
            plt.title(f"{oname} por tipo de transfusão (transfundidos)")
            plt.xlabel("tipo transfusão")
            plt.ylabel(oname)
            if SHOW_PLOTS:
                plt.show()
            plt.close()


# =========================
# 8) RASTREIOS (Tabela 2)
# =========================
print_section("8) RASTREIOS (Tabela única n e %)")

if not screening_cols:
    print_warn("Nenhuma coluna de rastreio encontrada pelos nomes/padrões comuns.")
else:
    rows = []
    # evitar duplicados: às vezes o mesmo rastreio foi achado por dois nomes
    used = set()
    for label, col in screening_cols.items():
        if col in used:
            continue
        used.add(col)
        b = to_binary_sim_nao(df[col])
        n_total = int(b.notna().sum())
        n_sim = int((b == 1).sum())
        pct = (n_sim / n_total * 100) if n_total else np.nan
        rows.append({"rastreio": label, "coluna": col, "n_total_com_dado": n_total, "n_SIM": n_sim, "pct_SIM": pct})

    rast_table = pd.DataFrame(rows).sort_values("pct_SIM", ascending=False)
    print(rast_table.to_string(index=False))


print_section("FIM")
print("Análise concluída. Se apareceu algum [AVISO], é porque o Excel não tem a coluna necessária com nome/padrão detectável.")
```
